```python
import pandas as pd
import os
import numpy as np
```

```python
sequence_lenght = 3
versions = {"Dreem": ["dodo", "dodh"], "SleepPhysionet": ["2018"]}
datasets = ["Dreem", "SleepPhysionet"]
model_paths = ["models/CCL/", "models/SCL/"]
model_name = ["chambon2018/", "tinysleepnet/"]
```

```python
# Initialize an empty DataFrame to store results
results = [pd.DataFrame() for _ in range(len(model_paths))]

# Loop over model directories
for i, model_path in enumerate(model_paths):
    for mn in model_name:
        for dataset in datasets:
            for v in versions[dataset]:
                result_path = "seqlen=%d/%s/%s/test_results.csv" % (
                    sequence_lenght,
                    dataset,
                    v,
                )
                result_path = model_path + mn + result_path

                df = pd.read_csv(result_path)
                df.pop("Unnamed: 0")
                df.pop("test_ck")
                df.pop("fold")
                mean = df.mean()
                std = df.std()

                results[i] = pd.concat(
                    [
                        results[i],
                        pd.DataFrame(
                            {
                                "model": mn[:-1],
                                "dataset": dataset,
                                "version": v,
                                "metric": df.columns,
                                "mean": mean,
                                "std": std,
                            }
                        ),
                    ]
                )

# Convert DataFrame to LaTeX table and write to file
for i in range(len(results)):
    with open(str(i) + "results.md", "w") as f:
        f.write(
            results[i]
            .set_index(["model", "dataset", "version", "metric"])
            .to_markdown()
        )

results[0].set_index(["model", "dataset", "version", "metric"]).head(n=20)
```
